# 확률의 종류

###1. 확률변수와 확률분포

####1) 확률변수

- 표본공간 내 각 사건들의 실수값을 대응시키는 함수
- Event(사건 등)의 수치적인 표현

####2) 확률분포

- 확률변수가 취할 수 있는 값과 그 값에 대한 가능성
- 이산형분포(discrete probability distribution) & 연속형 분포(continuous probability distribution)
- 확률질량함수(probability mass function, PMF) & 확률밀도함수(probability density function, PDF)

###2. 확률질량함수와 확률밀도함수

####1) 확률질량함수 특징

- P(X - $x_{\rm i}$) $\dashrightarrow$ X라는 값이 $x_{\rm i}$ 라는 특정값을 가질 확률
- 0 $\le$ P(x) $\le$ 1, i = 1,2,3,... n
- $\sum_{i=1}^n$P($x_{\rm i}$) = 1

####2) 확률밀도함수 특징

- $f(x)\ge0$, $-\infty\lt x \lt \infty$
- $\int_{i=1}^nf(x_{\rm i})dx = 1$
- 임의 수 c, d에 대해 $P(c\le x \le d) = \int_c^df(x)dx$

####3) 확률질량함수와 확률밀도함수의 차이점

- $f(x)$는 확률 X, 1초과가 가능하다.
- 확률변수 X가 c와 d사이에서 취할수 있는 값들은 $f(x)$의 면적
- $P(x = a) = 0 \dashrightarrow P(a \le x \le b) = p(a \lt x \lt b) = p(a \le x \lt b)$  

###3. 분포의 종류

####1) 누적분포함수(Cumulative Distribution Function, CDF)

- 어떤 확률 분포에 대해 확률 변수가 특정값보다 작거나 같은 확률 {$X \le x$}
- $F(x) = P(X \le x)$ : 확률변수 X의 누적분포함수
- 이산형과 연속형에 동일하게 적용가능
- $F(x)$는 $x$에 대한 비감소 함수(감소하지 않음)
- $\lim_{x\rightarrow\infty}F(x) = F(\infty)  = 1$
- $\lim_{x\rightarrow\infty}F(x) = F(-\infty) = 0$


####2) 결합확률분포(Joint Probability Distribution)

- 두개 이상의 변수가 확률적 관계를 가지고 관측된 경우의 대응확률
- Marginal probability 주변확률, 결합확률분포에서 하나의 사건에 대한 확률
- $P(X = x | Y = y) = \frac{P(X=x, Y=y)}{P(Y=y)} = \frac{P(x, y)}{P{\rm y}(Y)}$
- $P(X=x, Y=y) = P(X=x) * P(Y=y)$ 

####3) 푸아송 분포 

- 이산형 분포
- 독립적인 사건이 발생하는 상한이 정해지지 않은 분포
- 예를들어 물고기에게 먹이를 주었을 경우, 물고기가 몇 마리 올까? $\rightarrow$ 물고기 "온다/안온다" 자체는 독립적인 사건이지만 몇마리나 올지에 대한 상한선은 정해지지 않았음.
- $\lambda$(람다) : 구간 내 평균 발생 횟수
- 푸아송 분포 확률 질량 함수 : $f(x) = \sum_{r=0}^xP(x = r)$
- 푸아송 분포 확률 질량 함수 평균 : $E(x) = \sum_{x=0}^{\infty}x\frac{\lambda*e^{-\lambda}}{x!}$
- 문제 $\rightarrow$ 아마존 한달 평균 30명이 피라냐 습격으로 죽는다. $\rightarrow$ 한 달 30명이 죽을 확률, 하루에 2건이상 사고발생

####4) 정규분포 

- 연속형 분포
- 확률변수 X 의 확률 밀도 함수가 $f(x) = \frac{1}{\root\of{2\pi\sigma}}*e\frac{-(x-m)^2}{2\sigma^2}$ (단, $-\infty < x < \infty$) 일때 X의 확률분포를 정규분포라하고 $N(m, \sigma^2)$
- 임의의 실수 x에 대해 $f(x) > 0$
- 곡선은 직선 x - m에 관해 대칭
- 곡선과 x축 사이의 넓이는 1이다.
- x의 확률 분포 평균은 m이고, 분산은 $\sigma^2$이다.
- X가 [a, b]에 속할 확률은 $P(a \le x \le b)$는 구간 [a, b]에서 곡선과 x축 사이의 넓이와 같음.

####5) 지수분포

- 어떤 독립적 사건이 푸아송 분포에 의해 발생할 때, 지정된 시점으로부터 이 사건이 일어 날때 까지의 걸린시간을 측정
- 지수함수 $f(x) = \lambda * e^{-\lambda x}$
- 평균, 분산 (생략)
- 문제 $\rightarrow$ 물고기가 평균 10분에 5번 입질, 첫번째 입질로부터 다음입질이 올때까지 걸린 시간 (평균 1분에 0.5번 입질, $\lambda$ = 0.5)
- $f(t) = 0.5e^{-0.5t}$, t>0인 분포를 따름

####6) 균등분포

- 연속확률 분포 중, 특정구간내의 값들이 나타날 가능성이 균등한 분포

####7) Z - 분포

- 표준정규분포
- 정규분포를 표준화 시킨 것
- $Z = \frac{X-\mu}{\sigma}$

####8) $\chi^2$분포

- $Z^2 = \frac{(X-\mu)^2}{\sigma^2}, \nu = 1$
- 자유도 : n - 1 $\rightarrow$ 1은 추정된 모수
- 표본 평균과 모평균이 같다고 가정할 수 있는 이유는 모평균의 불편측정치(unbiased estimated value) 때문
- 모평균과 표본평균이 같다고 가정하기 때문에 n번째 표본은 나머지 표본이 정해지면 자동으로 정해짐, 따라서 자유도는 n-1

####9) t - 분포

- W.S.Gosset
- 표준정규분포를 따르는 확률변수를 독립인 $\chi^2/\nu$의 제곱근으로 나눈 확률 변수
- t ~$\frac{N(0,1)}{\root\of{\chi^2/\nu}}$
- 모집단 확률 변수 X가 정규분포를 따르고 그에 따라 표본집단이 정규분포를 따르는 경우 $\frac{\bar X - \mu}{S/\root\of{n}}$이 식은 자유도 n-1인 t분포를 따른다.
